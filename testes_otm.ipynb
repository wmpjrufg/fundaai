```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle


def plot_elementos_fundacao(df_res, label_col="Elemento"):
    """
    Desenha os elementos de fundação no plano cartesiano
    a partir de df_res (saída do obj_teste).

    Parâmetros
    ----------
    df_res : DataFrame
        Deve conter:
        - 'xg (m)', 'yg (m)'
        - 'h_x (m)', 'h_y (m)'
    label_col : str, opcional
        Nome da coluna para identificação dos elementos
    """

    fig, ax = plt.subplots(figsize=(9, 6))
    ax.set_aspect("equal", adjustable="box")
    ax.set_xlabel("x (m)")
    ax.set_ylabel("y (m)")
    ax.grid(True, linestyle="--", alpha=0.4)

    for _, r in df_res.iterrows():
        x_inf = r["xg (m)"] - r["h_x (m)"] / 2
        y_inf = r["yg (m)"] - r["h_y (m)"] / 2

        sapata = Rectangle(
            (x_inf, y_inf),
            r["h_x (m)"],
            r["h_y (m)"],
            fill=False,
            edgecolor="black",
            linewidth=1.5,
        )

        ax.add_patch(sapata)

        if label_col in df_res.columns:
            ax.text(
                r["xg (m)"],
                r["yg (m)"],
                str(r[label_col]),
                ha="center",
                va="center",
                fontsize=11,
                weight="bold",
            )

    # limites automáticos
    margem = max(df_res["h_x (m)"].max(), df_res["h_y (m)"].max())
    ax.set_xlim(
        df_res["xg (m)"].min() - margem,
        df_res["xg (m)"].max() + margem,
    )
    ax.set_ylim(
        df_res["yg (m)"].min() - margem,
        df_res["yg (m)"].max() + margem,
    )

    return fig, ax

fig, ax = plot_elementos_fundacao(df_res)
```
