```python
#tensão resistente pelo método Urbano r. Alonso
def sigma_adm_alonso(nspt: float)-> float:
    '''
    Esta função calcula a tensão resistente do solo a partir do nspt médio das três cotas a partir da cota de assentamento
    válida para nspt's até 20
    
    args:
        nspt (float): valor do nspt médio das três cotas a apartir da cota de assentamento
    retunrs:
        float: valor da tensão admissível do solo
    '''
    return nspt/30 # em mPa
```

```python
#tensão resistente pelo método cintra albieiro
def sigma_adm_cintra_albieiro(nspt: float, q: float)-> float:
    '''
    Esta função calcula a tensão resistente do solo a partir do nspt médio das três cotas a partir da cota de assentamento
    válida para nspt's até 20

    args:
        nspt (float): valor do nspt médio das três cotas a apartir da cota de assentamento
        q (float): valor da carga do solo acima da cota de assentamento
    retunrs:
        float: valor da tensão admissível do solo
    '''
    return nspt/50 + q
```
