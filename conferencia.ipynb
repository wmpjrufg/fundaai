```python
n_comb = 3
df = pd.read_excel('teste_wand.xlsx')
df = tensao_adm_solo(df)
df
```

```python
h_x = 0.8
h_y = 0.6
max_tensoes = []
min_tensoes = []

# Verificando a tensão em todos os itens de fundação
for idx, row in df.iterrows(): 
    t_max_aux = []
    t_min_aux = []

    # Checando para cada combinação
    for i in range(1, n_comb + 1):
        aux = f'c{i}'
        t_max_val, t_min_val = calcular_sigma_max(
                                                    row[f'Fz-{aux}'],
                                                    row[f'Mx-{aux}'],
                                                    row[f'My-{aux}'],
                                                    h_x,
                                                    h_y
                                                 )
        t_max_aux.append(t_max_val)
        t_min_aux.append(t_min_val)
        
    # Armazenando os valores máximos e mínimos
    max_tensoes.append(max(t_max_aux))
    min_tensoes.append(min(t_min_aux))

# Adiciona as colunas ao df
df['max_tensao'] = max_tensoes
df['min_tensao'] = min_tensoes

# Checagem tensão mínima
df['resultado'] = np.where(
                            df['min_tensao'] < 0,
                            np.abs(df['min_tensao'] / df['sigma_adm (kPa)']),    # Comportamento se valores negativos
                            df['min_tensao'] / df['sigma_adm (kPa)']             # Comportamento se valores >= 0
                          )

# Checagem tensão máxima
df['resultado'] = np.where(
                            df['max_tensao'] < 0,
                            df['max_tensao'] / df['sigma_adm (kPa)'],    # Comportamento se valores negativos
                            df['max_tensao'] / df['sigma_adm (kPa)']             # Comportamento se valores >= 0
                          )
```
