```python
#transformacao da planilha_padrao pro modelo teste_wand (com 1 header)

def padrao_to_flat(df_padrao):
    novo_df=pd.DataFrame()
    n_comb=0
    for col in df_padrao.columns:
        if"combinação" in col[0] and col[1] in ['Fz', 'Mx', 'My']:                 #so funcionara caso esteja escrito "combinação" dessa maneira, ainda sem tratar variacoes de possiveis erros de usuários
            comb_indice=col[0].strip().split()[-1]                  #"combinação X" tiro os espacos em branco inicial e final(caso tenha) com strip e divido ela com split e pego a ultima parte que no caso é o indice
            nome_col= f"{col[1]}-c{comb_indice}"
            n_comb=int(comb_indice)    #pega a qtd de combinacoes, o ultimo indice vai ser o numero de combinacoes q é o q vai ficar salvo em n_comb, pra tratar erros de usuario como combinacoes fora de ordem, usar set()
        else:
            nome_col=f"{col[0]}"
        novo_df[nome_col]=df_padrao[col]

    return novo_df, n_comb
df, n_comb=padrao_to_flat(df_p)

df
```

```python
def calcular_sigma_max(f_z: float, m_x: float, m_y: float, h_x: float, h_y: float) -> tuple[float, float]:
    """
    Esta função determina a tensão máxima e a tensão mínima na fundação rasa do tipo sapata

    Args
    f_z: Carregamento na direção z, da combinação mais desfavorável (kN)
    m_x: Momento em x da combinação mais desfavorável (kN.m)
    m_y: Momento em y da combinação mais desfavorável (kN.m)

    Returns
    sigma_max: Tensão máxima que age na sapata (kPa)
    sigma_min: Tensão minima que age na sapata (kPa)

    """
    
    m_x = abs(m_x)
    m_y = abs(m_y)
    sigma_fz = f_z / (h_x * h_y)
    aux_mx = 6 * (m_x / f_z) / h_x
    aux_my = 6 * (m_y / f_z) / h_y
    
    return (sigma_fz) * (1 - aux_mx - aux_my), (sigma_fz) * (1 + aux_mx + aux_my)
```

```python
# determinando a combinação mais desfavorável
t_sd = []
for _, row in df.iterrows():
    sig = []
    for i in range(1, n_comb+1):
        aux = f'c{i}'
        t_min_aux, t_max_aux = calcular_sigma_max(row[f'Fz-{aux}'], row[f'Mx-{aux}'], row[f'My-{aux}'], 0.6, 0.6)
        sig.append((t_min_aux, t_max_aux))
    t_sd.append(sig)
t_sd
```
