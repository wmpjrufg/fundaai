```python

def build_dataset_from_population(pop2d: np.ndarray, df_template: pd.DataFrame,
                                  n_comb: int, f_ck: float, h_z: float,
                                  obj_func) -> pd.DataFrame:
    """
    Converte uma população (n_pop x 2N) em dataset
    colunas: hx0..hx(N-1), hy0..hy(N-1), of
    """
    n_population, n_dimensions = pop2d.shape
    N = len(df_template)
    assert n_dimensions == 2 * N, f"Esperado 2*len(df) = {2*N}, veio {n_dimensions}"

    hx_cols = [f"hx{i}" for i in range(N)]
    hy_cols = [f"hy{i}" for i in range(N)]
    cols = hx_cols + hy_cols + ["of"]

    rows = []
    for i in range(n_population):
        x = pop2d[i, :]

        of, df_novo = obj_func(x, args=(df_template, n_comb, f_ck, h_z))
        row = list(x[:N]) + list(x[N:]) + [of]
        rows.append(row)

    return pd.DataFrame(rows, columns=cols)

N = len(df)
x_lower = [0.6] * (2 * N)
x_upper = [10.0] * (2 * N)

n_population = 2000
seed = 42
pop2d = initial_population_01(
    n_population=n_population,
    n_dimensions=N*2,
    x_lower=x_lower,
    x_upper=x_upper,
    seed=seed,
    use_lhs=True
)
pop2d = np.asarray(pop2d, dtype=float)
print(pop2d.shape)

dataset_df = build_dataset_from_population(
    pop2d=pop2d,
    df_template=df,
    n_comb=n_comb,
    f_ck=f_ck,
    h_z=h_z,
    obj_func=obj_teste
)

dataset_df.head()
```
