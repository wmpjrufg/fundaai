```python
dataset = {'hx0':[], 'hy0':[], 'hx1':[], 'hy1':[], 'hx2':[], 'hy2':[], 'hx3':[], 'hy3':[], 'hx4':[], 'hy4':[], 'of':[]}
for i in range(n_population):
    x = pop2d[i, :]
    of, df_novo = obj_teste(x, args=(df, n_comb, f_ck, h_z))
    dataset['hx0'].append(df_novo.loc[0, 'h_x (m)'])
    dataset['hy0'].append(df_novo.loc[0, 'h_y (m)'])
    dataset['hx1'].append(df_novo.loc[1, 'h_x (m)'])
    dataset['hy1'].append(df_novo.loc[1, 'h_y (m)'])
    dataset['hx2'].append(df_novo.loc[2, 'h_x (m)'])
    dataset['hy2'].append(df_novo.loc[2, 'h_y (m)'])
    dataset['hx3'].append(df_novo.loc[3, 'h_x (m)'])
    dataset['hy3'].append(df_novo.loc[3, 'h_y (m)'])
    dataset['hx4'].append(df_novo.loc[4, 'h_x (m)'])
    dataset['hy4'].append(df_novo.loc[4, 'h_y (m)'])
    dataset['of'].append(of)
```
